# Using Deepgram

## Transcribing a Local File

```python
import requests

headers = {
    'Content-Type': 'audio/wav',
}

data = open('male.wav', 'rb').read()
response = requests.post('https://brain.deepgram.com/v2/listen',
                         headers=headers, data=data, auth=('gabrielseemann@gmail.com', 'M2FW93fHkf9eyVa'))
```

```python
# response.text
jsponse = response.json()
```
